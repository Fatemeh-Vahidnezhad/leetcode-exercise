is Palindrome?

```python
def is_Palindrome(a):
    j = len(a)
    i = 0
    while i< j:
#         print('i',i)
#         print('j',j)
#         print(a[i], a[j-1])
        if a[i].lower() != a[j-1].lower():
            return 'it is not a polindrome word!'
        i += 1
        j -= 1
    return 'polindrome'

is_Palindrome('radar')
```

```python
def is_Palindrome2(a):
    j = len(a)
    i = 0
    while i< j:
        while i < j and not a[i].isalnum():
            i += 1

        while i<j and not a[j-1].isalnum():
            j -= 1

        if a[i].lower() != a[j-1].lower():
            return 'it is not a polindrome word!'

        i += 1
        j -= 1
    return 'polindrome'

```

```python
def titletoNumber(s):
    result = 0
    n = len(s)
    for i in range(n):
        result = 26 * result
        result += (ord(s[i])-ord('A') + 1)
    return result
print(titletoNumber('AAA'))
```

Anagram

s = 'listen'
t = 'silent'
every charactor in s can find in t

```python
def anagram(s,t):
    if len(s)== len(t):
        for i in range(len(s)):
            if s[i].lower() not in t.lower():
                return False
        return 'Anagram'
    return 'The number of strings are not equal!'
s = 'LISTEN'
t = 'silent'
anagram(s,t)
```

```python
def anagram2(s,t):
    if len(s) != len(t):
        return False
    return sorted(s.lower()) == sorted(t.lower())
s = 'LISTEN'
t = 'silent'
anagram2(s,t)
```
