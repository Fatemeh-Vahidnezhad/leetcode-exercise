remove digit and return bigger number


```python
def remove_digit (n, digit):
  lst = []
  lst1=[]
  for i in range(0, len(n)):
    while n[i] == digit:
      lst.append(n[:i] + n[i+1:])
      i+=1
  # covert string to int:
  for i in lst:
    lst1.append(int(i))
  return max(lst1)
```

```python
n = '1234267'
digit = '2'
remove_digit(n, digit)
```

Is palindrom?

```python
def palindrom(s):
  a = list(s)
  n = len(a)
  for i in range(0, n):
    if a[i] == a[n-i-1]:
      return True
    else:
      return False

print(palindrom('radar'))
```

```python
def recursive_S(s):
  def h(low, high):
    if low < high:
      s[low], s[high] = s[high], s[low]
      h(low +1 , high - 1)
  return h(0, len(s)-1)
a = 'dadasc'
s = list(a)
recursive_S(s)
print(s)
```

Two Sum

```python
def TwoSum(s, sum):
  for i in range(len(s)):
    for j in range(i):
      if s[i]+s[j] == sum:
        print(s[i], s[j])

s = [0,1,4,6,7,4,2,5]
TwoSum(s,7)
```

Pascal's Triangle\
1\
1 1\
1 2 1\
1 3 3 1\
1 4 6 4 1\
15 10 10 5 1\
\

```python
def PascalTriangle(n):   # n is the number of rows
  lst = []
  for r in range(1, n):
    row = [None for i in range(1, n+1)]
    row[0] = row[-1] =1
    for j in range(1, n-1):
      row[j]= lst[r-1][j-1] + lst[r-1][j]
  lst.append(row)
  print(lst)
PascalTriangle(3)
```

SumZero\
n = 1 -> [0]\
n = 2 -> [-1,1]\
n = 3 -> [-1,1,0]\
n = 4 -> [-2,-1,1,2]\
n = 5 -> [-2,-1,1,2,0]\
n = 6 -> [-3,-2,-1,1,2,3]\
n = 7 -> [-3,-2,-1,1,2,3,0]\



```python
def SumZero(n):
  lst = []
  if n % 2 == 0:
    a = n/2
    for i in range(int(a)):
      lst.append(i+1)
      lst.append(-1*(i+1))
    return lst
  else:
    a = n/2
    for i in range(int(a)):
      lst.append(i+1)
      lst.append(-1*(i+1))
    lst.append(0)
    return lst


SumZero(7)
```

sign of product of an array:\
x = [0,1,4,5]--> 0\
x = [1,-1,-2,4] -- > +1\
x = [-1,6,4,5,3] --> 1\


```python
def arraySign(x):
  a = 1
  if 0 in x:
    return 0
  for i in x:
    a *= i
    # print(a)
  if a<0:
    return -1
  return 1
x = [10,-1,-3,-5]
arraySign(x)
```

second solution

```python
x = [-10,-1,-3,-5]
def arraySign2(x):
  for i in x:
    if i == 0:
      return 0
  if sum(i<0 for i in x)%2:
    return -1
  return 1
arraySign2(x)
```

Sublist with sum of digit

```python
x = [3,1,2,7,4,2,2,1,5]
digit = 9
def find_sublist(x, digit):
  lst = []
  for i in x:
    lst.append(i)
    while sum(j for j in lst) > digit:
      lst.remove(lst[0])
    if sum(j for j in lst) == digit:
      print(lst)
```

```python
find_sublist(x,digit)
```

buy stock on minimum  and buy it on max day, each number shows the price of stock on every day

```python
x = [7,1,3,0,5,6,9,12,3]
def MaxProfit(x):
  max_profit = 0
  min = float('inf')
  for i in x:
    if i <= min:
      min = i    # time of buying stock
  for j in range(x.index(min)+1, len(x)):
    profit = x[j]-min
    if profit > max_profit:
      max_profit = profit
      max_rev = x[j]
      date = x.index(max_rev)
  return min, max_profit, max_rev, date+1
MaxProfit(x)
```
